# 문제1-6) '지역별' 컬럼을 '시도'와 '구군'으로 구분해서 별도의 컬럼을 만드세요

````python
def get_province(x):
    temp=x.split(' ')
    sido=temp[0]
    if len(sido) !=4:
        return(sido[0:2]) 
    else:
        return(sido[0]+sido[2])
part_col=part_col.copy() # Warning 제거, 메모리는 더 먹지만 안전해짐
part_col['시도'] = part_col['지역명'].dropna().apply(get_province)
part_col['구군'] = part_col['지역명'].dropna().apply(lambda x: x.split(' ')[1])

part_col
````

# 문제1-7) '지역고용통계' 를 위한 전처리 함수를 작성하세요.

````python
def graduate_preprocrssing(path):
    read_path = pd.read_excel(path, skiprows=1)
    select_col=read_path[read_path['산업별'] == '전산업']
    part_col=select_col[['지역별','전체종사자']]
    part_col.columns=['지역명','고용자수']
#     part_col['고용자수'] = pd.to_numeric(part_col['고용자수']) Warning 생김
    
    def get_province(x):
        temp=x.split(' ')
        sido=temp[0]
        if len(sido) !=4:
            return(sido[0:2]) 
        else:
            return(sido[0]+sido[2])
    
    part_col=part_col.copy()
    part_col['시도'] = part_col['지역명'].dropna().apply(get_province)
    part_col['구군'] = part_col['지역명'].dropna().apply(lambda x: x.split(' ')[1])
    
    return part_col
````

# 문제2-5) '주민등록세대수' 전처리 함수 만들기

````python
def graduate_preprocrssing2(path):
    read_path2=pd.read_excel(path)
    read_path2.columns=['시도','구군','세대수']
    
    big_col = []
    for num, temp in enumerate(read_path2['시도']):
        if pd.isna(temp) :
            big_col.append(big_col[num-1]) # nan을 칸을 강제로 다 채움
        else:
            big_col.append(temp)

    def get_province(x):
        if len(x) !=4:
            return(x[0:2]) 
        else:
            return(x[0]+x[2])

    small_col=[]
    for i in big_col:
        small_col.append(get_province(i))

    read_path2['시도']=small_col
    select_path2=read_path2[read_path2['구군'] != '소계']

    return select_path2
````
