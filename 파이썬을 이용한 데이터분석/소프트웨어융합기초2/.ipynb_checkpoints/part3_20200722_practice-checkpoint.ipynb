```python
df_cafe = df_seoul[df_seoul["상호명_소문자"].str.contains('스타벅스|starbucks|이디야|이디아|ediya')].copy()
df_cafe['상호명_소문자'].unique()
```

```python
df_cafe.loc[df_cafe["상호명_소문자"].str.contains('스타벅스|starbucks'), "브랜드명"] = "스타벅스"
df_cafe
```

```python
df_cafe.loc[~df_cafe["상호명_소문자"].str.contains('스타벅스|starbucks'), "브랜드명"] = "이디야"
df_cafe
```

```python
df_cafe = df_cafe[df_cafe["상권업종중분류명"] == "커피점/카페"].copy()
df_cafe.shape

df_cafe["브랜드명"].value_counts()
```

```python
sns.countplot(data=df_cafe, x="브랜드명")
```

```python
df_seoul_star=df_seoul[df_seoul['상호명']=='스타벅스'].copy()
df_seoul_star[['위도','경도']].plot.scatter(x='경도',y='위도')
```

```python
# 중앙 좌표를 위해 평균을 구함
import folium
```

```python
lat=df_seoul_star['위도'].mean()
long=df_seoul_star['경도'].mean()
lat, long
```

```python
#lo까지 치면 tap누르셈. 자동완성 기능
m=folium.Map(location=[lat,long],tiles='Stamen Toner',zoom_start=12)
```

```python
for i in df_seoul_star.index:
    circle_lat= df_seoul_star.loc[i,'위도']
    circle_long=df_seoul_star.loc[i,'경도']
    folium.CircleMarker(location=[circle_lat, circle_long], radius=5, color='green').add_to(m)
```
